Una vez tenemos la vista de todos los datos, vamos a empezar a manipularlos.

```python
df['is_4wd'] = df['is_4wd'].fillna(0)
df['is_4wd'] = df['is_4wd'].astype(bool)
df.info()
print(df['is_4wd'].unique())
print(df['is_4wd'].value_counts())
```

Ahora, vamos a rellenar los colores de nan como 'Unknown' 

```python
print(df['paint_color'].unique())
df['paint_color'] = df['paint_color'].fillna('unknown')
print(df['paint_color'].unique())
```

```python
print(df['cylinders'].unique())
df['cylinders'] = df['cylinders'].fillna('unknown')
print(df['cylinders'].unique())
```

```python
df.info()
```

```python
print(df['model'].unique())
df['brand'] = df['model'].str.split().str[0]
print(df['brand'].unique())
```

```python
print(df['odometer'].unique())
df['odometer'] = df['odometer'].fillna('unknown')
```

```python
df['model_year'] = df['model_year'].fillna('unknown')
print(df['model_year'].unique())
```
